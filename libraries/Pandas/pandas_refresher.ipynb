Series - A labeled 1D array

```python
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
```

SERIES

( loc uses labels, iloc uses position)

```python
age = [25, 30, 22]

s = pd.Series(age)
print("Default Series:\n", s)

s = pd.Series(age, index=["aaa", "bbb", "ccc"])
print("\nCustom Index:\n", s)

print("\nloc:", s.loc["aaa"])
print("iloc:", s.iloc[1])
```

DATAFRAME CREATION

```python
df = pd.DataFrame({
    "name": ["aaa", "bbb", "ccc"],
    "age": [25, 30, 22],
    "cgpa": [8.5, 9.0, 8.0]
})

print(df)
```

ACCESS DATA

```python
print("Column age:\n", df["age"])
print("\nMultiple columns:\n", df[["name", "age"]])
print("\nRow loc:\n", df.loc[2])
print("\nRow iloc:\n", df.iloc[1])
```

SET INDEX

```python
df2 = df.set_index("name")
print(df2)

print("\nAccess with index:\n", df2.loc["aaa"])
```

ADD COLUMN

```python
df["medium"] = ["english", "english", "french"]
print(df)
```

ADD ROW

```python
new_row = pd.DataFrame([
    {"name": "ddd", "age": 28, "cgpa": 8.2, "medium": "french"}
])

df = pd.concat([df, new_row], ignore_index=True)
print(df)
```

DISPLAY

```python
print(df.to_string())
print(df["name"].to_string())
```

SELECTION

```python
print(df.loc[0:2, ["name", "age"]])
```

FILTERING

```python
print(df.shape)
print(len(df))
```

```python
df["house_no"] = [10, 50, 60, 70]

filtered = df[
    (df["cgpa"] > 8) &
    (df["house_no"] > 40) &
    (df["name"] == "ddd")
]

print(filtered)
```

AGGREGATIONS





```python
print("Sum:\n", df.sum(numeric_only=True))
print("\nMean:\n", df.mean(numeric_only=True))
print("\nCount:\n", df.count())
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMin:\n", df.min(numeric_only=True))
print("\nIndex of Min:", df["age"].idxmin())

print("\nColumn mean:", df["age"].mean())
```

GROUPBY

```python
df["gender"] = ["F", "F", "F", "F"]

grouped = df.groupby("gender")["age"].mean()
print(grouped)
```

MISSING VALUES

```python
df.loc[1, "age"] = np.nan

print("With NaN:\n", df)
print("\nCheck:\n", df.isna())

print("\nDrop:\n", df.dropna())
print("\nDrop subset:\n", df.dropna(subset=["age"]))

print("\nFill:\n", df.fillna({"age": 44}))
```

REPLACE

```python
df["age"] = df["age"].replace({25: 26})
print(df)
```

TYPE CONVERSION

```python
print(df.to_string)
df["age"] = df["age"].astype("float64")
print(df.dtypes)
```

DUPLICATES

```python
df2 = pd.concat([df, df.iloc[[0]]], ignore_index=True)

print(df2)
print(df.drop_duplicates())
```

STRING OPS

```python
df["name"] = df["name"].str.upper()
print(df)
```

Information

```python
print("Shape:", df.shape)

df.info()

print(df.describe())

print(df["name"].value_counts())
```
